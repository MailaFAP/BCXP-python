```python
# variáveis do tipo string
nome = 'João da Silva'
cidade = 'São Paulo'
cpf = '123.456.789-00'

print(nome.upper(), cidade.upper(), cpf.upper())
print(nome.lower(), cidade.lower(), cpf.lower())
print(nome.find('ã'), cidade.find('ã'), cpf.find('ã'))
print(len(nome), len(cidade), len(cpf))
print(cpf.replace('.', '').replace('-', ''))
```

```python

numero = '127957'
#Utilize uma variável auxiliar, soma, para acumular o somatório dos valores.
soma = 0
#Utilize o laço de repetição for … in; para percorrer cada caractere da string;
for caractere in numero:
  #Utilize a conversão do tipo string para o tipo inteiro (int(caractere)) para converter os caracteres em valores numéricos;
  soma += int(caractere)

print(f'O somatário dos caracteres da string {numero} é: {soma}')
```
